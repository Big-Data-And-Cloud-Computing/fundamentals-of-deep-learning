```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
```

```python
df = pd.DataFrame()
df['x'] = np.linspace(500,1000,num=20)
df['y'] = 2 * df.x + 5
df.head(10)
```

```python
fig, ax = plt.subplots(figsize=(3,2.5))

ax.scatter(df.x, df.y, s=15, color='#F46C43')
ax.plot(df.x, df.y, lw=.75, color='black')
ax.set_xlim(500,1000)
ax.set_ylim(0,2100)
ax.set_xlabel("Square feet")
ax.set_ylabel("Rent")
plt.title("Apt rent prices")

plt.show()
```

$y = \fbox{m} x + \fbox{b}$

or multiple

```python
# brute force model parameter space search
for m in np.linspace(0,3,num=10):
    for b in np.linspace(0,10,num=10):
        y = m * df.x + b
        mse = np.sum((y - df.y)**2)
        print(f"{m:4.1f},{b:4.1f} => MSE {mse:.1f}")
```
